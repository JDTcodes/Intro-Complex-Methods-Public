```python
#Define function for convoltuion operation. 

def conv2(x,f):
  #Input arguments are x an image, and f a convolution filter / kernel.
    x2 = np.zeros(x.shape)
    #We define x2 as a matrix with all zeroes with same shape as matrix x. 
    #Next we will fill in all the values of matrix x2 with new values according to the below define operation, we run at every i row and j column of x2. 
    for i in range(1,x.shape[0]-1): #interate over all i rows in range of 1 through the first dimension in the shape of x "minus 1".
        for j in range(1,x.shape[1]-1): #interate over all j columns in range of 1 through the second dimension in the shape of x "minus 1".
            #Note in python, range(a,b), used above, goes from a to b-1.
            #We cannot run the 3x3 convolution matrix on the edges of the image as the operation would be undefined. We can alleviate this problem with padding, but here we have no padding. We can run the operation through the entire image except we cannot run on the last pixel near any edge. 
            #We start at 1 instead of 0 due to lack of padding.
            #x.shape[0] gives number of rows, since python starts at 0, we would run this to x.shape[0]-1, but since we have no padding, we run to x.shape[0]-2.Using the range through (x.shape[1]-1) really runs through (x.shape[1]-1)-1=x.shape[1]-2, so that is good for no padding. 
            #
            #Wrap the lines below with \
            x2[i,j] = f[0,0] * x[i-1,j-1]  \
            +         f[0,1] * x[i-1,j]    \
            +         f[0,2] * x[i-1,j+1]  \
            +         f[1,0] * x[i,j-1]    \
            +         f[1,1] * x[i,j]      \
            +         f[1,2] * x[i,j+1]    \
            +         f[2,0] * x[i+1,j-1]  \
            +         f[2,1] * x[i+1,j]    \
            +         f[2,2] * x[i+1,j+1]
            #The above code states that "for all cells in the 2d image x", where each cell has its own value for i row and j column, we do the following. 
            #The convolution operation takes the "value in the" first row, first column of convolution matrix f and multiplies it by i-1 row, j-1 column of the image x.
            #We add to that the product of first row, second column of convolution matrix times (i-1) row, j column of the image x.
            #We add to that the product of first row, third column of convolution matrix times (i-1) row, (j+1) column of the image x.
            #We add to that the product of second row, first column of convolution matrix times (i) row, (j-1) column of the image x.
            #We add to that the product of second row, second column of convolution matrix times (i) row, (j) column of the image x.
            #We add to that the product of second row, third column of convolution matrix times (i) row, (j+1) column of the image x.
            #We add to that the product of third row, first column of convolution matrix times (i+1) row, (j-1) column of the image x.
            #We add to that the product of third row, second column of convolution matrix times (i+1) row, (j) column of the image x.  
            #We add to that the product of third row, third column of convolution matrix times (i+1) row, (j+1) column of the image x.
            #This summation yields the new value we store in the x2[i,j], the ith row and jth column of our new convolved image x2. 
    return x2
    #Once we iterate for all i and j, we return the new convolved image / feature map, x2. 
```

```python
#Convolution with padding.
#Define convolution operation function with padding, so we can run convolution operation on the edges of the matrix.
#For a 3x3 matrix, this requires us to append values to one additional row at the top and bottom, as well as one addiitonal column at left and right, of image. 

def conv2(w,f): #GPU conv with padding.
    n = conv2d(w.type(torch.int),f.type(torch.int)) 
    #Use previously define conv2d function to convolve for over w, and change data types of w and f to int. 
    n = pad(n, (1, 1, 1, 1)) #add ones to the sides of the matrix.
    #Add padding to sides of matrix so convolution defined on edges.
    #We will see these 1s as black cells when we plot the binary images below. 
    return n
```

```python
#Game of Life Rules:

#1.) Any live cell with two or three live neighbors survives.
#2.) Any dead cell with three live neighbors becomes a live cell.
#3.) All other live cells die in the next generation. Similarly, all other dead cells stay dead.
```

```python
#Here, we play the game!

for i in range(steps):
  #This sets up a loop to repeat operations below for number of desired steps, steps=1000 here.

    n = conv2(w,f) #matrix count neighbors.
    #This is convolution of f(3x3 matrix with all 1s except 0 in center) over the matrix w (20x20 random 1 or 0).
    #This part of the loop counts the neighbors for a given cell in the matrix for each. 
    #This will give us a matrix n, which contains a value for each cell in 20x20 matrix, where the value is the count of neighbors
    #See above example for n = conv2(w,f).
    
    w = (w * (n==2)) + (n==3) #board state w matrix based on board state w and count of neighbors per n matrix.
    #--->Programming in parallel to implement the rules for game of life at each step!
    #== yields True when True and False when False.
    #w is the matrix state of the board for a given step. It shows all the cells that are on (1) and off (0). 
    #
    #(w * (n==2)) says for each cell of w, if cell is True (a 1) and it has exactly 2 neighbors, them leave the cell True (a 1). Part of rule 1 above. 
    #(n==3) says if the cell has exactly 3 neighbors, then cell is True (a 1). Part of rule 1 and rule 2 above. 
    #All other cells die as they fail to be true for above noted conditions on w cells. Rule 3.

    A[i] = w
    #New w - board layout/state saves in A[i] for ith step.
    #Note A[0] is updated after w takes one step. Thus, A[0] is not the initial condition of board, but is the board after one step. 
  
#Loop takes steps(1000 here) many steps from 0 to (steps-1) (0 to 999 here). 
#Steps=1000 w state of board matrices recorded in A[0] through A[999].

#w0 saved prior is initial state of board.
#A[0] is board after one step. 
#A[999] is board after 1000 steps (final configuration).
```

```python
#Define custom animation function.

def make_ani(A, colormap='gray'):
  #Define animation function with input A (here a tensor of 1000 20x20 matrices). 
  #Sets colormap to grayscale. 

    fig, ax = plt.subplots()
    #Set variables fig, ax to equal plt.subplots()
    #plt.subplots() provides a way to plot multiple plots on a single figure. 
    #Given the number of rows and columns, it returns a tuple ( fig , ax ), giving a single figure fig with an array of axes ax.
    
    im = ax.imshow(A[0,:,:], cmap = colormap);
    #Display data as an image, i.e., on a 2D regular raster.
    #Takes input rows and columns of given A[0] matrix. 
    #Sets cmap = colormap, prior set colormap='gray'. 

    ax.axis('off')
    #Turns of the axis values. 
    
    fig.set_size_inches(6, 6)
    #The set_size_inches() method figure module of matplotlib library is used to set the figure size in inches.
    #We set this for fig (the figure).

    def animate(data, im):
      #Define animate function with argument inputs of data (see below) and im (see above). 
        im.set_data(data)
        #Runs set_data on data(see below) for im.
        #This sets the data for im as data.
    
    def step():
        #Define step function with no input. 
        for i in range(A.shape[0]):
          #Sets to loop step over range of A.shape[0]=1000 steps. 
            data = A[i,:,:]
            #For each i in loop, we set all rows and column for that A[i] equal to data. 
            yield data
            #When the Python yield statement is hit, the program suspends function execution and returns the yielded value to the caller.
            #We yield the data (see above for data).

    return animation.FuncAnimation(fig, animate, step, interval=100, repeat=True, fargs=(im,))
    #Makes an animation by repeatedly calling a function func animate for picture, then step t go to next data. 
    #Reference fig we set size for above. 
    #Inteval = 100 is the delay in milliseconds between consecutive animation runs. 
    #Repeat true is whether the animation repeats when the sequence of frames is completed. here yes. 
    #fargs=(im,) is for additional arguments to pass to each call to function, here we pass im. 
```

```python
make_ani(A)
#Call animation function on A, so it will display steps of game stored in A[i], where i takes 0 through 999 (1000 matrices).
```
